### Syntaxis

**Definition of functions:**  
Python instead of many languajes does not use {} for functions or any kind of ifs, instead it uses **:** at the end of the *if/for/def* statemens and uses tabs to index each start and finish.   

**End of functions and data-types**  
There are no semicolons **;** to endlines on this languaje and there is no need to set the data type before hand.   
To ensure a function ended, just make sure the indentation is below the function on the same level on the start def and we leave a empty line.  
For example, we dont need to set int number = 20, we just do number = 20 and thats it.  

**Comments:**   
We can do single comments using **#**, or a multiline using **'''TEXT'''**


### Best Practices
**Naming:**   
In python we must name things as this 5 cases:
- Variables, Functions, Methods, Packages, Modules : *lower_case_with_underscores*
- Classes and Exceptions: *CapWords*
- Protected Methods and Internal Functions: *_single_leading_underscord(self, ...)*
- Private Methods: *__double_leading_underscore(self, ...)*
- Constants: *ALL_CAPS_WITH_UNDERSCORES*

We should avoid one letter vars, they must be self explanatorial (We can do it on short blocks).
```python
for e in elements:
    e.test()
```
Also when naming, we should use the elements name first example:

elements = ...   
elements_active = ...  
elements_defunct = ...  

**Indentation:**  
We should set a consistent amount of indentation to make sure we are on the correct line.  
Google good practices uses a set value of 2 spaces per tab, making it always the same no matter the IDE.  

**Boolean comparision:**  
As we do on C++ or any other languaje, we should not use *a == True* to check if a boolean is indeed true, we should simply let the value itself be checked.  

**List Comprenhension:**  
We should use list comprehension when possible:
```python
a = [1,2,3]
b = []
for i in a:
    if i > 2:
        b.append(i)
```
This can be rewritten as:
```python
a = [1,2,3]
b = [i for i in a if i > 2]
```
**With keyword:**  
The usage of *with* makes sure that when finishing using a file it gets closed.
```python
with open('file.txt') as f:
    do_some
```
**Imports:**  
We should import all the module, not simple parts. 
**Use of comments:**  
The code must be self explanatory so we don't need to use comments on every part of the code to explain simple things.

### Loops

Python has two loops, while and for.

- **continue**: stops the current iteration and continues to the next


We can only access the items through a for loop.

Functions:

To join them we use: 
```python
cars.union(cars2)
```

### Object Oriented Programming 

OOP lets us create classes and objects which helps us develop a clear structure, makes it easy to maintain and makes reusable apps with less code.  

**Classes and Objects**  
Classes defines what an object should look like, while an object is created based on the class.  

To create a class we should use the keyword *class*:
```python
class MyClass: 
    x = 5
```
To create an objet we use MyClass() like this:
```python
p1 = MyClass()
print(p1.x)
```
This should print 5.

**The \_\_init\_\_() Function**  
This function is always excecuted when initializing the class, this assings values to objects properties, like the constructor in Java.  
```python
class Person: 
    def __init__(self, name, age):
        self.name = name
        self.age = age

p1 = Person("Jhon", 36)
```

**The \_\str\_\_() Function**  
This function controls what should be returned if the object is represented as a string. Like the ToString method in Java.
```python
class Person: 
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def __str__(self):
        return f"{self.name}({self.age})"
        
p1 = Person("Jhon", 36)
```

**Object Methods:**  
Objects can have methods, those are functions that belong to the object.  
They are called inside the class and are used by the name of the method.
```python
class Person: 
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def myFunc(self):
        print("Hello my name is " + self.name)
        
p1 = Person("Jhon", 36)
p1.myFunc()
```

**Self Parameter**  
The self parameter is a reference to the current instance of the class, and is used to access variables that belong to the class.  
We can call it whatever we want but it is standarized that we use self.  

**Modifications:**  
We can modifiy directly them for example using *p1.age = 40*, also we can delete properties from the object by using *del* keyword or delete the object at all.  


**Inheritance**  
This allows us to define a class that inherits all methods and properties from another class.  
We have a **Parent Class** and a **Child Class**, this makes child inherit properties from parent class.  
Example:  
```python
#Parent Class: 
class Person:
  def __init__(self, fname, lname):
    self.firstname = fname
    self.lastname = lname

  def printname(self):
    print(self.firstname, self.lastname)

#Child Class
class Student(Person):
  pass
```
We use pass when we dont want to add any new properties or methods to the class.  
There Student class inherits init constructor and the printname function from Person.  

**Iterators**  
We cam oterate through data using an iterator, an iterator is an object that contains a number of values.  
It instead of the init or str, uses **\_\_iter\_\_()** and **\_\next\_\_()** methods.  
We create them by using *iter()* function and move using *next()* like this:
```python
mytuple = ("Apple", "Orange", "Banana")
myit = iter(mytuple)

print(next(myit))
print(next(myit))
print(next(myit))
```

This should get printed all the elements in the tuple.  

**Polymorphism**  

This means that there are various methods/functions/operators that have the same name and can be executed on many objects or classes, for example we have *len()* function.  

This let us overwrite functions or methods when we inherit some method from another class.

#### Series
Series is a one dimensional array holding data of any type.  
To create a serie we can use the following command:  

````python
import pandas as pd
a = [1, 7, 2]
myvar = pd.Series(a)
````

We can create labels for the array, and call the values by its index or the label (if we had set one).

````python
print(myvar[0])
myvar = pd.Series(a,index=["x","y","z"])
print(myvar["y"])
````

We can create a series from a dictionary, this set the dict keys as labels.  
Also, if we want to create a series with some items from the dictionary we can use the **index** value and specify which ones we want to use.
